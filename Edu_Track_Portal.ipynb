## ✅ Step 3 — Generate Datasets (100 Students, 20 Faculty, Admin)

This cell generates synthetic student and faculty data, along with admin credentials. It uses `numpy.random` for data generation and `pandas` to create the DataFrames.

```python
import numpy as np

np.random.seed(42)

# ─────────────────────────────────────────────
# BRANCHES AND SUBJECTS
# ─────────────────────────────────────────────
BRANCHES = ['CSE', 'ECE', 'AIDS', 'AIML', 'IT', 'MECH', 'CIVIL']
YEARS    = ['1st Year', '2nd Year', '3rd Year', '4th Year']

SUBJECTS = {
    'CSE':   ['Data Structures', 'DBMS', 'OS', 'CN', 'Software Engg'],
    'ECE':   ['Signals', 'Electronics', 'Microprocessors', 'VLSI', 'Communications'],
    'AIDS':  ['ML', 'Deep Learning', 'Statistics', 'Data Mining', 'Big Data'],
    'AIML':  ['AI Fundamentals', 'NLP', 'Computer Vision', 'Reinforcement Learning', 'Python'],
    'IT':    ['Web Tech', 'Cloud Computing', 'Cyber Security', 'IoT', 'Mobile Computing'],
    'MECH':  ['Thermodynamics', 'Fluid Mechanics', 'Machine Design', 'Manufacturing', 'CAD'],
    'CIVIL': ['Structures', 'Surveying', 'Concrete Technology', 'Geotechnical', 'Hydrology']
}

# ─────────────────────────────────────────────
# GENERATE 100 STUDENTS
# ─────────────────────────────────────────────
first_names = ['Akhil','Vasu','Naseer','Sudeep','Ravi','Priya','Sai','Anjali','Rahul','Divya',
               'Kiran','Sneha','Arjun','Pooja','Manoj','Lakshmi','Vijay','Swathi','Arun','Kavya',
               'Harish','Meena','Suresh','Nandini','Ramesh','Deepika','Ganesh','Pavithra','Chandu','Bhavya',
               'Naveen','Sirisha','Prasad','Madhuri','Rajesh','Sunitha','Mohan','Geetha','Vinay','Jyothi',
               'Srikanth','Padma','Naresh','Rekha','Krishna','Uma','Santosh','Ratna','Dinesh','Lavanya',
               'Teja','Hema','Lokesh','Sravani','Venkat','Mounika','Ashok','Naga','Praveen','Swetha',
               'Ramu','Sridevi','Balaji','Tulasi','Raju','Anitha','Gopal','Vani','Madhu','Sailaja',
               'Pavan','Sudha','Rohit','Yamini','Satish','Revathi','Kishore','Sowmya','Anil','Pushpa',
               'Navin','Theja','Uday','Chaitra','Prem','Shirisha','Srinivas','Aparna','Vivek','Himaja',
               'Ajay','Keerthi','Sunil','Sushma','Tarun','Bhanu','Vishal','Radha','Charan','Jhansi']

last_names = ['Kumar','Reddy','Rao','Sharma','Naidu','Gupta','Verma','Singh','Patel','Nair']

students = []
for i in range(100):
    branch     = BRANCHES[i % len(BRANCHES)]
    year       = np.random.choice(YEARS)
    attendance = round(np.random.uniform(55, 98), 1)
    marks      = {sub: int(np.random.normal(65, 15)) for sub in SUBJECTS[branch]}
    marks      = {k: max(0, min(100, v)) for k, v in marks.items()}
    avg_marks  = round(np.mean(list(marks.values())), 2)
    cgpa       = round(min(10, max(0, 4 + (avg_marks - 40) * 0.09 + (attendance - 55) * 0.03 + np.random.normal(0, 0.3))), 2)
    sid        = f'23NA1A{str(500 + i).zfill(4)}'
    name       = first_names[i] + ' ' + np.random.choice(last_names)
    email      = f'{first_names[i].lower()}{500+i}@edutrack.edu'
    password   = str(500 + i)
    total_fee  = 75000
    paid_fee   = int(np.random.choice([0, 25000, 37500, 50000, 75000]))
    due_fee    = total_fee - paid_fee

    row = {'StudentID': sid, 'Name': name, 'Email': email, 'Password': password,
           'Branch': branch, 'Year': year, 'Attendance': attendance,
           'AvgMarks': avg_marks, 'CGPA': cgpa,
           'TotalFee': total_fee, 'PaidFee': paid_fee, 'DueFee': due_fee}
    row.update(marks)
    students.append(row)

df_students = pd.DataFrame(students)

def cgpa_band(c):
    if c >= 9.0:   return 'Outstanding'
    elif c >= 7.5: return 'Excellent'
    elif c >= 6.0: return 'Good'
    elif c >= 5.0: return 'Average'
    else:          return 'At Risk'

df_students['CGPABand'] = df_students['CGPA'].apply(cgpa_band)
df_students['PassFail'] = df_students['CGPA'].apply(lambda x: 'Pass' if x >= 5.0 else 'Fail')

# ─────────────────────────────────────────────
# GENERATE 20 FACULTY
# ─────────────────────────────────────────────
fac_names = ['Dr. Siva Rama Krishna','Prof. Durga Bhavani','Dr. N. Satyanarayana',
             'Prof. Anand Rao','Dr. Priya Sharma','Prof. Venkat Reddy',
             'Dr. Lakshmi Devi','Prof. Suresh Kumar','Dr. Rajesh Naidu','Prof. Kavita Patel',
             'Dr. Arun Verma','Prof. Sneha Gupta','Dr. Mohan Singh','Prof. Swathi Nair',
             'Dr. Ganesh Rao','Prof. Deepika Reddy','Dr. Harish Kumar','Prof. Meena Sharma',
             'Dr. Ravi Naidu','Prof. Madhuri Gupta']

faculty = []
for i, name in enumerate(fac_names):
    branch  = BRANCHES[i % len(BRANCHES)]
    fid     = f'FAC{str(100 + i).zfill(3)}'
    email   = f'fac{100+i}@edutrack.edu'
    passw   = f'fac{100+i}'
    # assign ~14 students per faculty proportionally
    assigned = df_students[df_students['Branch'] == branch]['StudentID'].tolist()[:8]
    faculty.append({'FacultyID': fid, 'Name': name, 'Email': email,
                    'Password': passw, 'Branch': branch,
                    'AssignedStudents': ','.join(assigned)})

df_faculty = pd.DataFrame(faculty)

# ─────────────────────────────────────────────
# ADMIN
# ─────────────────────────────────────────────
ADMIN = {'email': 'admin@edutrack.edu', 'password': 'admin123', 'name': 'Administrator'}

print(f'✅ Dataset Generated!')
print(f'   👨‍🎓 Students : {len(df_students)}')
print(f'   👨‍🏫 Faculty  : {len(df_faculty)}')
print(f'   🔑 Admin    : 1')
print(f'\n📌 Sample Students:')
display(df_students[['StudentID','Name','Branch','Year','CGPA','CGPABand','Attendance','PassFail']].head(10))
```

### 🔐 Login Functions

```python
def admin_login(email, password):
    """Verifies admin credentials."""
    if email == ADMIN['email'] and password == ADMIN['password']:
        print(f"Admin login successful! Welcome, {ADMIN['name']}.")
        return True
    else:
        print("Invalid Admin credentials.")
        return False

def faculty_login(email, password):
    """Verifies faculty credentials."""
    faculty_user = df_faculty[(df_faculty['Email'] == email) & (df_faculty['Password'] == password)]
    if not faculty_user.empty:
        print(f"Faculty login successful! Welcome, {faculty_user['Name'].iloc[0]}.")
        return True
    else:
        print("Invalid Faculty credentials.")
        return False

def student_login(student_id, password):
    """Verifies student credentials."""
    student_user = df_students[(df_students['StudentID'] == student_id) & (df_students['Password'] == password)]
    if not student_user.empty:
        print(f"Student login successful! Welcome, {student_user['Name'].iloc[0]}.")
        return True
    else:
        print("Invalid Student credentials.")
        return False

print('✅ Login functions created.')
```

```python
# Test valid faculty login
print('Testing valid faculty login:')
valid_faculty_email = df_faculty['Email'].iloc[0] # Get email of the first faculty member
valid_faculty_password = df_faculty['Password'].iloc[0] # Get password of the first faculty member
faculty_login(valid_faculty_email, valid_faculty_password)

print('\nTesting invalid faculty login:')
# Test invalid faculty login
invalid_faculty_email = 'nonexistent@edutrack.edu'
invalid_faculty_password = 'wrongpassword'
faculty_login(invalid_faculty_email, invalid_faculty_password)
```

### 🚀 Interactive Login

```python
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

# --- Admin Login Widgets ---
admin_email_input = widgets.Text(description='Email:')
admin_password_input = widgets.Password(description='Password:')
admin_login_button = widgets.Button(description='Admin Login', button_style='success')
admin_output = widgets.Output()

def on_admin_login_clicked(b):
    with admin_output:
        clear_output()
        if admin_login(admin_email_input.value, admin_password_input.value):
            display(HTML("<p style='color: green;'>Logged in as Admin!</p>"))
        else:
            display(HTML("<p style='color: red;'>Invalid Admin credentials.</p>"))

admin_login_button.on_click(on_admin_login_clicked)

admin_box = widgets.VBox([
    widgets.HTML("<h4>👨‍💼 Admin Login</h4>"),
    admin_email_input,
    admin_password_input,
    admin_login_button,
    admin_output
])

# --- Faculty Login Widgets ---
faculty_email_input = widgets.Text(description='Email:')
faculty_password_input = widgets.Password(description='Password:')
faculty_login_button = widgets.Button(description='Faculty Login', button_style='info')
faculty_output = widgets.Output()

def on_faculty_login_clicked(b):
    with faculty_output:
        clear_output()
        if faculty_login(faculty_email_input.value, faculty_password_input.value):
            display(HTML("<p style='color: green;'>Logged in as Faculty!</p>"))
        else:
            display(HTML("<p style='color: red;'>Invalid Faculty credentials.</p>"))

faculty_login_button.on_click(on_faculty_login_clicked)

faculty_box = widgets.VBox([
    widgets.HTML("<h4>👨‍🏫 Faculty Login</h4>"),
    faculty_email_input,
    faculty_password_input,
    faculty_login_button,
    faculty_output
])

# --- Student Login Widgets ---
student_id_input = widgets.Text(description='Student ID:')
student_password_input = widgets.Password(description='Password:')
student_login_button = widgets.Button(description='Student Login', button_style='danger')
student_output = widgets.Output()

def on_student_login_clicked(b):
    with student_output:
        clear_output()
        if student_login(student_id_input.value, student_password_input.value):
            display(HTML("<p style='color: green;'>Logged in as Student!</p>"))
        else:
            display(HTML("<p style='color: red;'>Invalid Student credentials.</p>"))

student_login_button.on_click(on_student_login_clicked)

student_box = widgets.VBox([
    widgets.HTML("<h4>👨‍🎓 Student Login</h4>"),
    student_id_input,
    student_password_input,
    student_login_button,
    student_output
])

# --- Display all login boxes ---
display(widgets.HBox([admin_box, faculty_box, student_box]))
```

```python
# @title AI prompt cell

import ipywidgets as widgets
from IPython.display import display, HTML, Markdown,clear_output
from google.colab import ai

dropdown = widgets.Dropdown(
    options=[],
    layout={'width': 'auto'}
)

def update_model_list(new_options):
    dropdown.options = new_options
update_model_list(ai.list_models())

text_input = widgets.Textarea(
    placeholder='Ask me anything....',
    layout={'width': 'auto', 'height': '100px'},
)

button = widgets.Button(
    description='Submit Text',
    disabled=False,
    tooltip='Click to submit the text',
    icon='check'
)

output_area = widgets.Output(
     layout={'width': 'auto', 'max_height': '300px','overflow_y': 'scroll'}
)

def on_button_clicked(b):
    with output_area:
        output_area.clear_output(wait=False)
        accumulated_content = ""
        for new_chunk in ai.generate_text(prompt=text_input.value, model_name=dropdown.value, stream=True):
            if new_chunk is None:
                continue
            accumulated_content += new_chunk
            clear_output(wait=True)
            display(Markdown(accumulated_content))

button.on_click(on_button_clicked)
vbox = widgets.GridBox([dropdown, text_input, button, output_area])

display(HTML("""
<style>
.widget-dropdown select {
    font-size: 18px;
    font-family: "Arial", sans-serif;
}
.widget-textarea textarea {
    font-size: 18px;
    font-family: "Arial", sans-serif;
}
</style>
"""))
display(vbox)
```

### 💰 Fee Status Summary for All Students
